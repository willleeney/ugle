```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_customgraph(G, clustering):

    degree_sequence = sorted((d for n, d in G.degree()), reverse=True)

    fig = plt.figure("Degree of a random graph", figsize=(8, 8))
    # Create a gridspec for adding subplots of different sizes
    axgrid = fig.add_gridspec(5, 4)

    ax0  = fig.add_subplot(axgrid[0:3, :])
    #Gcc = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])
    pos = nx.spring_layout(G)
    nx.draw_networkx_nodes(G, pos, ax=ax0, node_size=20,
                           node_color=clustering)
    nx.draw_networkx_edges(G, pos, ax=ax0, alpha=0.4)
    ax0.set_title("Graph: ")
    ax0.set_axis_off()

    ax1 = fig.add_subplot(axgrid[3:, :2])
    ax1.plot(degree_sequence, "b-", marker="o")
    ax1.set_title("Degree Rank Plot")
    ax1.set_ylabel("Degree")
    ax1.set_xlabel("Rank")

    ax2 = fig.add_subplot(axgrid[3:, 2:])
    ax2.bar(*np.unique(degree_sequence, return_counts=True))
    ax2.set_title("Degree histogram")
    ax2.set_xlabel("Degree")
    ax2.set_ylabel("# of Nodes")

    fig.tight_layout()
    plt.show()


# Create graph
num_clusters = 2
num_nodes = 100
G = nx.gnp_random_graph(num_nodes, 0.02)

# Assign clusters to nodes
node_clusters = np.random.randint(0, high=num_clusters, size=(num_nodes))
cmap = plt.get_cmap('Set1').colors
clustering = [cmap[node] for node in node_clusters]

plot_customgraph(G, clustering)
```

```python
# Create graph
num_clusters = 3
num_nodes = 500
G = nx.watts_strogatz_graph(num_nodes, k=3, p=0.2)

# Assign clusters to nodes
node_clusters = np.random.randint(0, high=num_clusters, size=(num_nodes))
cmap = plt.get_cmap('Set1').colors
clustering = [cmap[node] for node in node_clusters]

plot_customgraph(G, clustering)
```
